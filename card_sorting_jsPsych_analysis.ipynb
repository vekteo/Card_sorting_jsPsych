```
import glob
import seaborn as sns; sns.set()
import matplotlib.pyplot as plt
import pandas as pd
import os
import numpy as np
from google.colab import files
```

```
# get data file names
filenames = glob.glob("*.csv")
dfs = []
for filename in filenames:
    dfs.append(pd.read_csv(filename, encoding='ISO-8859-1'))

# Concatenate all data into one DataFrame
big_frame = pd.concat(dfs, ignore_index=True)
big_frame
```

```
#leave only the important columns
df = big_frame.loc[:, ['Participant Public ID', 'STAT_nr_of_trials','STAT_p_of_correct_trials','STAT_nr_of_perseverative_responses','STAT_nr_of_perseverative_errors','STAT_nr_of_non_perseverative_errors','STAT_p_perseverative_errors','STAT_nr_of_total_errors','STAT_p_of_errors','STAT_category_achieved','STAT_p_of_conceptual_level_responses','STAT_failure_to_maintain_set','STAT_trials_to_complete_first_category']]
df.head()
```

```
#drop instruction, feedback trials
df = df.dropna(subset=['STAT_nr_of_trials'])
df.head()
```

```
#save data
df.to_csv("card_sorting.csv",index=False)
files.download("card_sorting.csv")
```
